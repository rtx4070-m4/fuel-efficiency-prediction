```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='muted')
%matplotlib inline
```

## 1. Load the Dataset

```python
df = pd.read_csv('../data/auto-mpg.csv')
print(f'Shape: {df.shape}')
df.head(10)
```

## 2. Basic Information

```python
df.info()
```

```python
df.describe()
```

## 3. Missing Values & Data Cleaning

The `horsepower` column contains `?` as placeholders for missing values.

```python
# Count '?' in horsepower
print('Missing horsepower entries:', (df['horsepower'] == '?').sum())

# Convert and drop NAs
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f'Shape after cleaning: {df.shape}')
```

## 4. Distribution of Target Variable (MPG)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df['mpg'], bins=25, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].set_xlabel('MPG')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of MPG')

axes[1].boxplot(df['mpg'], vert=True)
axes[1].set_ylabel('MPG')
axes[1].set_title('Box Plot of MPG')

plt.tight_layout()
plt.show()

print(f'Mean MPG: {df["mpg"].mean():.2f}')
print(f'Median MPG: {df["mpg"].median():.2f}')
print(f'Std MPG: {df["mpg"].std():.2f}')
```

## 5. Distribution of Predictor Variables

```python
numeric_cols = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, col in zip(axes.flatten(), numeric_cols):
    ax.hist(df[col], bins=20, edgecolor='black', alpha=0.7, color='teal')
    ax.set_title(col.title())
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 6. Correlation Heatmap

```python
corr = df[['mpg'] + numeric_cols].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r', center=0,
            linewidths=0.5, square=True)
plt.title('Pearson Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 7. Scatter Plots: Features vs MPG

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, col in zip(axes.flatten(), numeric_cols):
    ax.scatter(df[col], df['mpg'], alpha=0.5, s=20, color='darkorange')
    ax.set_xlabel(col.title())
    ax.set_ylabel('MPG')
    ax.set_title(f'{col.title()} vs MPG')
plt.tight_layout()
plt.show()
```

## 8. MPG by Origin

Origin: 1 = USA, 2 = Europe, 3 = Japan

```python
origin_labels = {1: 'USA', 2: 'Europe', 3: 'Japan'}
df['origin_label'] = df['origin'].map(origin_labels)

plt.figure(figsize=(8, 5))
sns.boxplot(x='origin_label', y='mpg', data=df, palette='Set2')
plt.title('MPG Distribution by Origin')
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.tight_layout()
plt.show()

print(df.groupby('origin_label')['mpg'].describe().round(2))
```

## 9. MPG Trend Over Model Years

```python
yearly = df.groupby('model year')['mpg'].mean()

plt.figure(figsize=(10, 5))
plt.plot(yearly.index, yearly.values, marker='o', linewidth=2, color='seagreen')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.title('Average MPG by Model Year')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 10. Pair Plot (Key Features)

```python
key_features = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
sns.pairplot(df[key_features], diag_kind='kde', plot_kws={'alpha': 0.4, 's': 15})
plt.suptitle('Pair Plot of Key Features', y=1.02)
plt.show()
```

## 11. Key Insights

1. **Weight is the strongest negative predictor** of MPG — heavier cars are less fuel-efficient.
2. **Displacement and horsepower** also show strong negative correlations with MPG.
3. **Model year has a positive relationship** with MPG — newer cars (in this dataset era) tend to be more efficient.
4. **Japanese cars** (origin 3) achieve the highest average MPG, followed by European cars.
5. **Cylinders** is strongly correlated with displacement and weight (multicollinearity), which motivates techniques like PCA and regularisation.
6. **Acceleration** has a weak positive correlation with MPG and may not be a strong standalone predictor.
7. The target variable (MPG) is roughly **right-skewed**, with a mean around 23 MPG.
